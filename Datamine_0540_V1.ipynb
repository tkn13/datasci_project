```python
import pandas as pd
from dateutil.relativedelta import relativedelta
import numpy as np

# 1. โหลดข้อมูล (delimiter=';')
df = pd.read_csv('Motor vehicle insurance data.csv', delimiter=';')

# 2. สร้างตัวแปรชั่วคราวสำหรับคำนวณ (เพื่อไม่ให้กระทบข้อมูลจริงในตาราง)
# ใช้ dayfirst=True เพื่อให้อ่านวันที่ถูก (วัน/เดือน/ปี)
temp_renewal = pd.to_datetime(df['Date_last_renewal'], dayfirst=True, errors='coerce')
temp_birth = pd.to_datetime(df['Date_birth'], dayfirst=True, errors='coerce')
temp_licence = pd.to_datetime(df['Date_driving_licence'], dayfirst=True, errors='coerce')

# 3. ฟังก์ชันคำนวณปี
def calculate_years_diff(end_date, start_date):
    if pd.isnull(end_date) or pd.isnull(start_date):
        return np.nan
    try:
        years = relativedelta(end_date, start_date).years
        if years < 0: return np.nan
        return years
    except:
        return np.nan

# 4. สร้างคอลัมน์ใหม่ต่อท้าย (old และ Driving_experience)
df['old'] = [calculate_years_diff(r, b) for r, b in zip(temp_renewal, temp_birth)]
df['Driving_experience'] = [calculate_years_diff(r, l) for r, l in zip(temp_renewal, temp_licence)]

# ==========================================
# ส่วนที่เพิ่ม: คำนวณ Length_of_vehicle_usage
# ==========================================
# ดึงปี ค.ศ. จาก temp_renewal (ที่คุณแปลงไว้แล้ว) มาลบกับ Year_matriculation
df['Length_of_vehicle_usage'] = temp_renewal.dt.year - df['Year_matriculation']

# 5. แสดงตัวอย่างแค่ 5 บรรทัดแรก
print("ตัวอย่างข้อมูล 5 แถวแรก (สังเกตคอลัมน์ขวาสุด):")
# ถ้าใช้ใน Jupyter/Colab ให้ใช้ display(df.head())
# ถ้ารันใน Python ธรรมดาใช้ print(df.head())

display(df.head())
```
